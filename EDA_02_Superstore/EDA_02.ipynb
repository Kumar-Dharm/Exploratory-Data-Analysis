```python
# 1.	What is the overall sales trend over time? Can you identify any seasonal patterns?

plt.figure(figsize=(8, 5))
#sns.barplot(df, x=df['Ship_Date'].dt.month, y=df['Sales'])
#sns.barplot(df, x=df['Ship_Date'].dt.quarter, y=df['Sales'])
sns.barplot(df, x=df['Ship_Date'].dt.year, y=df['Sales'])
plt.title('Overall Sales Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Define the time period
periods = ['year', 'quarter', 'month']

# Create subplots for each period
fig, axes = plt.subplots(len(periods), 1, figsize=(12, 8))

for i, period in enumerate(periods):
    # Group the data by the specified period and calculate the sum of sales
    if period == 'year':
        df_new = df.groupby(df['Ship_Date'].dt.year)['Sales'].sum().reset_index()
        x_label = 'Year'
    elif period == 'quarter':
        df_new = df.groupby(df['Ship_Date'].dt.to_period('Q'))['Sales'].sum().reset_index()
        x_label = 'Quarter'
    elif period == 'month':
        df_new = df.groupby(df['Ship_Date'].dt.to_period('M'))['Sales'].sum().reset_index()
        x_label = 'Month'
    
    # Create a bar plot for the current period
    sns.barplot(data=df_new, x=df_new['Ship_Date'], y=df_new['Sales'], ax=axes[i])
    
    axes[i].set_title(f'Sales Trend by {period.capitalize()}')
    axes[i].set_xlabel(x_label)
    axes[i].set_ylabel('Sales')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3.	What are the top-selling categories and sub-categories of products by sales volume and revenue?

# Category

# Group by 'Category' and sum up the sales and count the number of records
category_sales = df.groupby('Category').agg({'Sales': 'sum', 'Category': 'count'}).rename(columns={'Category': 'Count'})

# Sort the categories by sales volume in descending order
category_sales = category_sales.sort_values(by='Sales', ascending=False)

# Display top-selling categories 
print('Top-Selling Categories by Sales Volume:')
category_sales
```

```python
# Sub_Category

# Group by 'Sub_Category' and sum up the sales and count the number of records
sub_category_sales = df.groupby('Sub_Category').agg({'Sales': 'sum', 'Sub_Category': 'count'}).rename(columns={'Sub_Category': 'Count'})

# Sort the sub_categories by sales volume in descending order
sub_category_sales = sub_category_sales.sort_values(by='Sales', ascending=False)

# Display top-selling sub_categories 
print('Top-Selling Sub_Categories by Sales Volume:')
sub_category_sales
```

```python
# Group by 'Category' and sum up the sales
category_sales = df.groupby('Category')['Sales'].sum().reset_index().rename(columns={'Sales': 'Total Sales'})

# Sort the categories by sales volume in descending order
category_sales = category_sales.sort_values(by='Total Sales', ascending=False)

# Display top-selling categories by sales volume
print('Top-Selling Categories by Sales Volume:')
category_sales
```

```python
# Now, let's find the top-selling sub-categories within each category
# Group by both 'Category' and 'Sub_Category' and sum up the sales
sub_category_sales = df.groupby(['Category', 'Sub_Category'])['Sales'].sum().reset_index().rename(columns={'Sales': 'Total Sales'})

# Sort the sub-categories by sales volume in descending order within each category
sub_category_sales = sub_category_sales.sort_values(by=['Category', 'Total Sales'], ascending=[True, False])

# Display top-selling sub-categories by sales volume within each category
print('Top-Selling Sub-Categories by Sales Volume within Each Category:')
sub_category_sales
```

```python
# Group by 'Sub_Category' within each 'Category' and do the same calculations
sub_category_sales = df.groupby(['Category', 'Sub_Category']).agg({'Sales': 'sum', 'Sub_Category': 'count'}).rename(columns={'Sub_Category': 'Count'})

# Sort the sub-categories by sales volume in descending order within each category
sub_category_sales = sub_category_sales.sort_values(by=['Category', 'Sales'], ascending=[True, False])

# Create a rank based on sales for each category
sub_category_sales['Rank'] = sub_category_sales.groupby('Category')['Sales'].rank(ascending=False, method='dense')

# Display top-selling sub-categories
print('Top-Selling Sub-Categories by Sales Volume:')
sub_category_sales
```

```python
# Group by 'Sub_Category' within each 'Category' and sum up the sales
sub_category_sales = df.groupby(['Category', 'Sub_Category'])['Sales'].sum().reset_index().rename(columns={'Sales': 'Total Sales'})

# Sort the sub-categories by sales volume in descending order within each category
sub_category_sales = sub_category_sales.sort_values(by=['Category', 'Total Sales'], ascending=[True, False])

# Create a rank based on sales for each category
sub_category_sales['Rank'] = sub_category_sales.groupby('Category')['Total Sales'].rank(ascending=False, method='dense')

# Display top-selling sub-categories with rank
print('Top-Selling Sub-Categories by Sales Volume with Rank:')
sub_category_sales
```

```python
# Calculate moving averages for different window sizes
window_sizes = [7, 30, 90]  # in days
for i in window_sizes:
    df[f'MA_{i}'] = df['Sales'].rolling(window=i).mean()

# Plot the sales data and moving averages
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Sales'], label='Actual Sales')
for i in window_sizes:
    plt.plot(df.index, df[f'MA_{i}'], label=f'Moving Avg ({i} days)')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.title('Sales Trend Over Time')
    plt.legend()
    plt.show()
```
